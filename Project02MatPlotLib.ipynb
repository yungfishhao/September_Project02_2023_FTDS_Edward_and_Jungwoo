```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
some_data = pd.read_csv("McDonalds_HK_Breakfast_Nutritional_Data.csv")
some_data.drop(columns=some_data.columns[0], inplace=True)
some_data
```

```python
items = [
     "hotcake-2pcs",
     "hotcakes-3pcs",
     "deluxe-breakfast",
     "hotcakes-with-sausage-hash-browns",
     "jumbo-breakfast",
     "grilled-chicken-twisty-pasta-original",
     "ham-n-egg-twisty-pasta-original",
     "sausage-n-egg-twisty-pasta-original",
     "grilled-chicken-twisty-pasta-tonkotsu",
     "ham-n-egg-twisty-pasta-tonkotsu",
     "sausage-n-egg-twisty-pasta-tonkotsu"
]

some_data.index = items
some_data
```

```python
some_data.dtypes
```

```python
breakfast_items = items
nutrients_by_grams = {
    'Protein (g)': some_data['Protein (g)'],
    'Total Fat (g)': some_data['Total Fat (g)'],
    'Saturated Fat (g)': some_data['Saturated Fat (g)'],
    'Trans Fat (g)': some_data['Trans Fat (g)'],
    'Carbohydrates (g)': some_data['Carbohydrates (g)'],
    'Sugar (g)': some_data['Sugars (g)']
}

x = np.arange(len(breakfast_items))  # the label locations
width = 0.1  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in nutrients_by_grams.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, visible=False)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Nutrients by gram')
ax.set_title('Nutrition Data by Meal')
ax.set_xticks(x + width, breakfast_items, rotation=90, fontsize=6)
ax.legend(loc='upper right', ncols=1)
ax.set_ylim(0, 100)

plt.show()
```

```python
fig, ax = plt.subplots()
bar_container = ax.bar(items, some_data['Energy (kcal)'])
ax.set(ylabel = 'Kcals', title = "Energy by Meal")
ax.set_xticks(x, breakfast_items, rotation=90, fontsize=8)
plt.show()
```

```python
fig, ax = plt.subplots()
bar_container = ax.bar(items, some_data['Sodium (mg)'])
ax.set(ylabel = 'milligrams', title = "Sodium by Meal")
ax.set_xticks(x, breakfast_items, rotation=90, fontsize=8)
plt.show()
```
